```python
import cv2
import numpy as np
import earthpy.plot as ep
```

```python
imgN = cv2.imread('/content/2021_E142983N.jpg')

# Extract the based on the bands
b_n, g_n, r_n = cv2.split(imgN)
img_array = np.array(r_n)

heat_index = (((-42.379 + (2.04901523 * img_array) + (10.14333127 * img_array) - (0.22475541 * img_array) - (6.83783e-3 * img_array) - (5.481717e-2 * img_array) + (1.22874e-3 * img_array) + (8.5282e-4 * img_array) - (1.99e-6 * img_array)))/272.15)+17.2222222

WS16 = np.power(8, 0.16)
Wind_Chill = 35.74 + (0.6215 * heat_index) - (35.75 * WS16) + (0.4275 * heat_index * WS16)
ep.plot_bands(Wind_Chill,cmap='Blues',title='Wind chill temperature generate from the image')
ep.plot_bands(heat_index,cmap='RdBu_r',title='Relative surface Temperature generate from the image')
```
